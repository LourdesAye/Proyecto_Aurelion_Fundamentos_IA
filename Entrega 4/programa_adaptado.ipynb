```python
# --- FUNCI√ìN SEMESTRE ---
def obtener_semestre(mes):
    return 1 if mes <= 6 else 2

# Agregar columna semestre
df_ventas["semestre"] = df_ventas["fecha_venta"].dt.month.apply(obtener_semestre)

# --- Agrupaci√≥n para obtener presencia por ciudad ---
presencia = (
    df_ventas.groupby(["id_producto", "semestre", "ciudad_encodear"])["id_venta"]
        .count()
        .reset_index(name="ventas")
)

# convertimos a 1/0
presencia["vendido"] = presencia["ventas"].apply(lambda x: 1 if x > 0 else 0)

# --- Crear matriz pivoteada: columnas por ciudad ---
matriz_ciudades = presencia.pivot_table(
    index=["id_producto", "semestre"],
    columns="ciudad_encodear",
    values="vendido",
    fill_value=0
).reset_index()

# --- Convertir todas las columnas de ciudades a int ---
cols_ciudades = matriz_ciudades.columns.drop(["id_producto", "semestre"])
matriz_ciudades[cols_ciudades] = matriz_ciudades[cols_ciudades].astype(int)

print(matriz_ciudades)
```

```python
# Mes con mayor venta por producto y semestre
# Agrupar por producto, semestre y mes
ventas_mes = (
    df_ventas.groupby(["id_producto", "semestre", "mes_venta"])["cantidad"]
    .sum()
    .reset_index()
)

# Obtener el mes con mayor venta por producto y semestre
mes_mayor_venta = (
    ventas_mes.sort_values(["id_producto", "semestre", "cantidad"], ascending=[True, True, False])
    .drop_duplicates(subset=["id_producto", "semestre"])
    .rename(columns={"mes_venta": "mes_max_venta",
        "cantidad": "cantidad_mes_max"})
)

print(mes_mayor_venta)
```

```python
promedios = (
    df_ventas.groupby(["id_producto", "semestre"])
        .agg(
            importe_promedio=("importe", "mean"),
            cantidad_promedio=("cantidad", "mean"),
            precio_promedio=("precio_unitario", "mean")
        )
        .reset_index()
)
promedios["importe_promedio"] = promedios["importe_promedio"].round(2) # si quer√©s mantener decimal
promedios["cantidad_promedio"] = promedios["cantidad_promedio"].round().astype(int)   
promedios["precio_promedio"] = promedios["precio_promedio"].round()
print(promedios)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Definir la variable Target (Y) , en este caso, rotaci√≥n
Y = df_final['rotacion']

# COLUMNAS A ELIMINAR
# Eliminar las que podrian causar Fuga de Datos (Leakage) y las que no son predictivas (IDs y constantes).
COLUMNAS_A_ELIMINAR = [
    'rotacion',             # Es el Target, no una Feature.
    'rotacion_semestral',   # Causa Leakage (componente que se utiliza para calcular la rotaci√≥n)
    'cantidad_dia_max',     # Causa Leakage 
    'cantidad_mes_max',     # Causa Leakage 
    'cantidad_promedio',    # Causa Leakage
    'semestre',             # Si es constante, no aporta valor.
    'id_producto',          # ID (no predictivo sin encoding avanzado).
    'frecuencia',          # causa Leakage
]

# Definir Features (X) eliminando las columnas problem√°ticas
X = df_final.drop(columns=COLUMNAS_A_ELIMINAR, errors='ignore')

print("Features (X) para el modelo: \n",X)
print("\nTarget (Y) para el modelo: \n",Y)

# revisar si frecuencia incluye cantidad 
```

```python
print(X.info())
print(X.head())
print(Y.info())
print(Y.head())
```

```python
# --- Divisi√≥n de Datos (Train/Test Split) ---
# Separamos 80% para entrenar y 20% para probar
# parte del modelo que se usa para entrenar y para evaluar el modelo
X_train, X_test, Y_train, Y_test = train_test_split(
    X,                  # datos de entrenamiento x
    Y,                  # datos a predecir
    test_size=0.20,     # tama√±o del test
    random_state=42,    # para reproducibilidad, que simpre sea igual
    stratify=Y          # mantener proporci√≥n de clases en ambos conjuntos, es decir, que haya balance entre rotaci√≥n alta y baja
)

print(f"\nConjunto de Entrenamiento (X_train): {len(X_train)} filas")
print(f"Conjunto de Prueba (X_test): {len(X_test)} filas")
print(f"Total filas: {len(X_train) + len(X_test)} filas")
print("Datos listos para el entrenamiento.")
```

```python
print(Y_train.info())
print(Y_test.info())
print(X_train.info())
print(X_test.info())
```

```python
# Se importan librer√≠as para modelos y evaluaci√≥n 
# Importa el clasificador de √°rbol de decisi√≥n, un modelo que divide el espacio de caracter√≠sticas en reglas if/else para clasificar.
from sklearn.tree import DecisionTreeClassifier           
# Importa el clasificador Random Forest, un ensamble de m√∫ltiples √°rboles de decisi√≥n que votan para mejorar robustez y reducir sobreajuste          
from sklearn.ensemble import RandomForestClassifier 
 # M√©tricas de evaluaci√≥n de modelos, como precisi√≥n y reporte de clasificaci√≥n                 
from sklearn.metrics import accuracy_score, classification_report    

# Inicializar
# Se crea un √°rbol de decisi√≥n 
    # con semilla fija para reproducibilidad 
    # y limita la profundidad m√°xima a 5 niveles para controlar el sobreajuste.
# Una semilla es un n√∫mero inicial que controla ese generador de n√∫meros aleatorios.
# Si fij√°s la semilla (ejemplo: random_state=42), el algoritmo siempre va a producir los mismos resultados 
# al volver a entrenar con los mismos datos.
dt_model = DecisionTreeClassifier(random_state=42, max_depth=5) # Decision Tree 

# Crea un bosque aleatorio con semilla fija y 100 √°rboles, un tama√±o com√∫n que equilibra rendimiento y costo computacional.
rf_model = RandomForestClassifier(random_state=42, n_estimators=100) # Random Forest

# Entrenar
# Ajusta el √°rbol de decisi√≥n usando las caracter√≠sticas X_train y las etiquetas Y_train.
dt_model.fit(X_train, Y_train)
# Ajusta el bosque aleatorio entrenando sus 100 √°rboles sobre (sub)muestras de X_train y Y_train.
rf_model.fit(X_train, Y_train)

# Predicci√≥n (se generar√°n predicciones sobre datos de prueba)
dt_pred = dt_model.predict(X_test)
rf_pred = rf_model.predict(X_test)

print("\n--- Resultados de Evaluaci√≥n ---")

# etiquetas reales (y_true), las predicciones del modelo (y_pred) y un nombre descriptivo (name).
def evaluate_and_report(y_true, y_pred, name):
    print(f"\n### {name} ###")
    print(f"Accuracy (Precisi√≥n Global): {accuracy_score(y_true, y_pred):.4f}") # se calcula precisi√≥n con accuracy , con 4 decimales
    # Genera e imprime el reporte con m√©tricas por clase (precision, recall, f1, soporte). 
        # target_names etiqueta las clases 0 y 1 con nombres legibles.
    # target_names[0] se asigna a la clase 0  y target_names[1] se asigna a la clase 1 por eso se pone as√≠
    print(classification_report(y_true, y_pred, target_names=['Alta Rotaci√≥n (0)', 'Baja Rotaci√≥n (1)']))

evaluate_and_report(Y_test, dt_pred, "√Årbol de Decisi√≥n")
evaluate_and_report(Y_test, rf_pred, "Bosque Aleatorio")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Pasos para mejorar el modelo
# 1. Definir el espacio de b√∫squeda (Grid)
# Elegimos hiperpar√°metros comunes del Random Forest para optimizar
param_grid = {
    # N√∫mero de √°rboles en el bosque
    'n_estimators': [50, 100, 200],
    # M√°xima profundidad de cada √°rbol (para controlar el overfitting)
    'max_depth': [3, 5, 8],
    # N√∫mero m√≠nimo de muestras requeridas para dividir un nodo interno
    'min_samples_split': [2, 5, 10]
}

# 2. Inicializar el modelo base
rf_model = RandomForestClassifier(random_state=42)

# 3. Inicializar GridSearchCV
# cv=5: Usamos 5-Fold Cross-Validation, esencial para datasets peque√±os.
# scoring='accuracy': La m√©trica que queremos maximizar.
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=5, 
    scoring='accuracy',
    verbose=1,
    n_jobs=-1 # Usa todos los n√∫cleos del procesador
)

# 4. Ejecutar la b√∫squeda en el conjunto de entrenamiento
print("Buscando la mejor combinaci√≥n de hiperpar√°metros...")
grid_search.fit(X_train, Y_train)

# 5. Obtener el mejor modelo y sus resultados
best_rf_model = grid_search.best_estimator_
best_score = grid_search.best_score_
best_params = grid_search.best_params_

print("\n--- Resultados de la Optimizaci√≥n ---")
print(f"Mejor Score (Accuracy CV): {best_score:.4f}")
print(f"Mejores Hiperpar√°metros: {best_params}")
```

```python
import joblib
import os
from sklearn.metrics import classification_report

# Pasos para guardar el modelo
# 1. Evaluaci√≥n en el Conjunto de Prueba
final_pred = best_rf_model.predict(X_test)

print("\n--- Evaluaci√≥n Final del Mejor Modelo ---")
print(classification_report(Y_test, final_pred, target_names=['Alta Rotacion (0)', 'Baja Rotaci√≥n (1)']))

# 2. Guardar el Mejor Modelo
# Definimos el nombre y la ruta del archivo
MODEL_FILE_NAME = 'best_random_forest_premium_predictor.joblib'
output_dir = 'modelos_guardados'

# Crear el directorio si no existe
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Guardar el modelo usando joblib (ideal para scikit-learn)
joblib.dump(best_rf_model, os.path.join(output_dir, MODEL_FILE_NAME))

print(f"\nModelo guardado exitosamente en: {os.path.join(output_dir, MODEL_FILE_NAME)}")

# Para cargar el modelo m√°s tarde, usar√≠as:
# loaded_model = joblib.load(os.path.join(output_dir, MODEL_FILE_NAME))
```
